```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10 data points
num_points = 10
x_values = np.random.uniform(-2, 2, num_points)
y_values = np.random.normal(x_values ** 2, 0.2, num_points)

# Construct the Vandermonde matrix
degree = num_points - 1 
V = np.vander(x_values, degree + 1, increasing=True)

# Perform linear least squares regression
coefficients, residuals, _, _ = np.linalg.lstsq(V, y_values, rcond=None)

# Plot the data points
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# Generate x values for the line
x_line = np.linspace(min(x_values) - 0.1, max(x_values) + 0.1, 1000)
# Compute y values for the line
y_line = x_line ** 2


# Compute the y values using the polynomial coefficients
y_curve = np.polyval(coefficients[::-1], x_line)

# Plot the polynomial curve
plt.plot(x_line, y_curve, color='red', label='Degree 9 Polynomial')

# Plot the line representing y = x^2
plt.plot(x_line, y_line, color='green', label='$y = x^2$')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example of an overfit function')
plt.grid(True)
plt.legend()
#plt.xlim(min(x_values) - 0.1, max(x_values) + 0.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10 data points
num_points = 5
x_values = np.random.uniform(-2, 2, num_points)
y_values = np.random.normal(x_values ** 2, 1, num_points)

# Construct the Vandermonde matrix
degree = num_points - 1 
V = np.vander(x_values, degree + 1, increasing=True)

# Perform linear least squares regression
coefficients, residuals, _, _ = np.linalg.lstsq(V, y_values, rcond=None)

# Print the coefficients
print("Coefficients of the polynomial:")
print(coefficients)

# Calculate the fitted values
y_fitted = np.dot(V, coefficients)

# Calculate the residual sum of squares (RSS)
RSS = np.sum((y_values - y_fitted) ** 2)
print("Residual sum of squares (RSS):", RSS)

# Plot the data points
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# Generate x values for the line
x_line = np.linspace(-2, 2, 100)
# Compute y values for the line
y_line = x_line ** 2

# Generate x values for the polynomial curve
x_curve = np.linspace(-2, 2, 1000)

# Compute the y values using the polynomial coefficients
y_curve = np.polyval(coefficients[::-1], x_curve)

# Plot the data points
plt.scatter(x_values, y_values, color='blue', label='Data Points')

# Plot the polynomial curve
plt.plot(x_curve, y_curve, color='red', label='Degree 9 Polynomial')

# Plot the line representing y = x^2
plt.plot(x_line, y_line, color='red', label='$y = x^2$')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data Points Generated from Distribution')
plt.grid(True)
plt.legend()
plt.show()
```
